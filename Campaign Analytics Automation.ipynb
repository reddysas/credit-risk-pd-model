```python
import pandas as pd

# Sample campaign data
data = pd.DataFrame({
    'channel': ['SMS', 'Email', 'SMS', 'Email'],
    'conversion_rate': [0.10, 0.20, 0.15, 0.25]
})

# Performance summary
report = data.groupby('channel').agg({
    'conversion_rate': 'mean'
}).reset_index()

print("Campaign Performance:\n", report)
```
